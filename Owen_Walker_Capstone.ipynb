## YT data

```python
with open('./data/ytdataset.json', "rb") as f: 
    ytd = json.load(f)
```

```python
ytd = pd.DataFrame(ytd)
```

```python
ytd.shape
```

```python
ytd.head()
```

```python
ytnames = pd.read_csv('./data/YouTube_Names.csv')
```

```python
ytnames.head()
```

```python
ytnames.shape
```

```python
ytnames.columns
```

```python
ytset = ytd.merge(ytnames, how = 'left', left_on = 'channelId', right_on = ' ID')
```

```python
ytset.info()
```

```python
ytset.isnull().sum()
```

```python
ytset = ytset.dropna()
```

```python
ytset.shape
```

```python
ytset.head()
```

```python
ytset = ytset.drop(columns = ['channelId', 'videoCategoryId', 'videoId', 'elapsedtime', 'videoPublished', 'likes/subscriber', 'views/subscribers',
                              'dislikes/views','comments/subscriber', 'dislikes/subscriber', 'totviews/totsubs', 'views/elapsedtime', 'comments/views'
                             , 'likes/dislikes', 'totalviews/channelelapsedtime', 'likes/views', 'totvideos/videocount'])
```

```python
ytset.head()
```

```python
ytset.shape
```

```python
ytset.to_csv('./data/ytset_cleaned.csv')
```
